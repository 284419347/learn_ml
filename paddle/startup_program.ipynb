```python
import paddle.fluid as fluid
import paddle
import numpy as np
from paddle.fluid.debugger import draw_block_graphviz
```

```python
new_startup = fluid.Program()
new_main = fluid.Program()

with fluid.program_guard(new_main, new_startup):
    image = fluid.layers.data(name='image', shape=[3, 24, 24], dtype='float32')
    predict = fluid.layers.fc(name='predict', input=image, size=1)
    draw_block_graphviz(new_main.blocks[0], path="predict.pdf")
    optimizer = fluid.optimizer.Adam(learning_rate=0.001)
    optimizer.minimize(predict)
    draw_block_graphviz(new_main.blocks[0], path="opt.pdf")
```
